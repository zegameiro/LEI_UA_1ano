b) Considerando a resist√™ncia do ar, calcule a energia mec√¢nica nos tr√™s instantes
ùë°_0=0, ùë°_1=0.4 s e ùë°_2=0.8 s.


```python
# -*- coding: utf-8 -*-
"""
Created on Wed Apr 21 17:00:09 2021

@author: cdazevedo
"""

#a)

import numpy as np
import matplotlib.pyplot as plt

dt=0.001
tf=1.00
n=np.int(tf/dt+0.1)

t=np.linspace(0,tf,n)


g=9.80      # m/s**2
vt=100*1000/3600  # m/s
vel0=100*1000/3600  # m/s
theta=10*np.pi/180  # rad
v0x=vel0*np.cos(theta)
v0y=vel0*np.sin(theta)
x0=0
y0=0



vel=np.empty(n)
y=np.empty(n)
vx=np.empty(n)
x=np.empty(n)
ax=np.empty(n)
ay=np.empty(n)
vx=np.empty(n)
vy=np.empty(n)
energia=np.empty(n)
vx[0]=v0x
vy[0]=v0y
x[0]=x0
y[0]=y0
massa=0.057

D=g/vt**2

for i in range(0,n-1):
    vel=np.sqrt(vx[i]**2+vy[i]**2)
    ax[i]=-D*vel*vx[i]
    ay[i]=-D*vel*vy[i]-g
    vx[i+1]=vx[i]+ax[i]*dt
    vy[i+1]=vy[i]+ay[i]*dt
    x[i+1]=x[i]+vx[i]*dt
    y[i+1]=y[i]+vy[i]*dt
    energia[i]=0.5*massa*vel**2+massa*g*y[i]
    if (i==0):
        stringPrint=str()+" "+str(energia[i])
        print ("T: {:.2f} s  E_m:{:.2f} J".format(t[i],energia[i]))
    if (t[i]<0.4 and t[i+1]>0.4):
        print ("T: {:.2f} s  E_m:{:.2f} J".format(t[i],energia[i]))
    if (t[i]<0.8 and t[i+1]>0.8):
        print ("T: {:.2f} s  E_m:{:.2f} J".format(t[i],energia[i]))
        
        
#Calcular o ultimo valor da energia mecanica:
vel=np.sqrt(vx[i+1]**2+vy[i+1]**2)
energia[i+1]=0.5*massa*vel**2+massa*g*y[i+1]




plt.figure()
plt.subplot(2, 1, 1)
plt.title(' Trajet√≥ria Bola de T√©nis com resist√™ncia do ar - Euler dt=0.001 s')
plt.ylabel('Y)')
plt.xlabel( 'x' )
plt.plot(x,y,'c')
plt.grid()
plt.subplot(2, 1, 2)

plt.title(' Energia Mec√¢nica (t) - Euler dt=0.001 s')
plt.ylabel('Em (J)')
plt.xlabel( 't (s)' )
plt.plot(t,energia,'c')
plt.grid()
plt.show()
```

c) Considerando a resist√™ncia do ar, calcule o trabalho realizado pela for√ßa de resist√™ncia do ar at√© √†s posi√ß√µes nos tr√™s instantes
ùë°_0=0, ùë°_1=0.4 s e ùë°_2=0.8 s.
Use a aproxima√ß√£o trapezoidal para calcular os integrais. A velocidade terminal da bola de t√©nis √© 100 km/h. AS massa da bola √© 57 g.


```python
# -*- coding: utf-8 -*-
"""
Created on Wed Apr 21 17:00:09 2021

@author: cdazevedo
"""

#a)

import numpy as np
import matplotlib.pyplot as plt

dt=0.001
tf=1.00
n=np.int(tf/dt)

t=np.linspace(0,tf,n+1)


g=9.80      # m/s**2
vt=100*1000/3600  # m/s
vel0=100*1000/3600  # m/s
theta=np.deg2rad(10) # rad
v0x=vel0*np.cos(theta)
v0y=vel0*np.sin(theta)
x0=0
y0=0


#Aqui uso n+1 por ser mais facil interpretar a formula da integra√ß√£o
vel=np.empty(n+1)
y=np.empty(n+1)
vx=np.empty(n+1)
x=np.empty(n+1)
ax=np.empty(n+1)
ay=np.empty(n+1)
vx=np.empty(n+1)
vy=np.empty(n+1)
energia=np.empty(n+1)
fun=np.empty(n+1)
aresx=np.empty(n+1)
aresy=np.empty(n+1)
int_trab_res=np.empty(n+1)

vx[0]=v0x
vy[0]=v0y
x[0]=x0
y[0]=y0
massa=0.057

D=g/vt**2



for i in range(0,n):	
    vel=np.sqrt(vx[i]**2+vy[i]**2)
    aresx[i]=-D*vel*vx[i]
    ax[i]=aresx[i]
    aresy[i]=-D*vel*vy[i]
    ay[i]=aresy[i]-g
    vx[i+1]=vx[i]+ax[i]*dt
    vy[i+1]=vy[i]+ay[i]*dt
    x[i+1]=x[i]+vx[i]*dt
    y[i+1]=y[i]+vy[i]*dt
    energia[i]=0.5*massa*vel**2+massa*g*y[i]
    #pede o trabalho da for√ßa resistiva, logo so interessa essa parte da acelera√ß√£o
    # W=integral(F dx)+integral(F dy)
    # dx √© dado por vx*dt e dy=vy*dt
    #Assim: INTEGRACAO   W=Integral(  Fx vx + Fy * vy) dt
    #For√ßa segundo X: segunda lei de Newton Fx=m*ax e Fy=m*ay (apenas as resistivas)
    
    fun[i]=massa*aresx[i]*vx[i]+massa*aresy[i]*vy[i]
    int_trab_res[i]=dt*((fun[0]+fun[i])*0.5+np.sum(fun[1:i]))    # NOTA fun[1:N] retorna um vetor de 1 a n-1, que √© oq ue queriamos. o python nos arrays √© exclusiv√©
    if (i==0):
        stringPrint=str()+" "+str(energia[i])
        print ("T: {:.2f} s  Trabalho resistivo:{:.2f} J".format(t[i],int_trab_res[i]))
    if (t[i]<0.4 and t[i+1]>=0.4):
        print ("T: {:.2f} s  Trabalho resistivo:{:.2f} J".format(t[i],int_trab_res[i]))
    if (t[i]<0.8 and t[i+1]>=0.8):
        print ("T: {:.2f} s  Trabalho resistivo:{:.2f} J".format(t[i],int_trab_res[i]))
        
 
#Calcular o ultimo valor da energia mecanica:
vel=np.sqrt(vx[i+1]**2+vy[i+1]**2)
energia[i+1]=0.5*massa*vel**2+massa*g*y[i+1]

#Calcular o ultimo valor do trabalho:
fun[i+1]=massa*aresx[i+1]*vx[i+1]+massa*aresy[i+1]*vy[i+1]
#na parte np.sum(fun[1:n]) e nao n-1, pois o python ao fazer 1:n o "n" √© exlusive
int_trab_res[i+1]=dt*((fun[0]+fun[n])*0.5+np.sum(fun[1:n]))
   
        

plt.figure()
plt.plot(t,int_trab_res,label='W Res')
# plt.plot(t,energia,label='W Res')
plt.ylabel('W_Res (J)')
plt.xlabel( 't (s)' )
plt.grid()

plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

dt=0.01
tf=10.00
n=np.int(tf/dt)

t=np.linspace(0,tf,n+1)


k= 1     # N/m
m = 1 #Kg 
w = np.sqrt(k/m)
x0 = 4
v0 = 0
a0 = 4
alpha = 0
A = x0

x_analitico= A*np.cos(w*np.arange(0,(n+1))*dt+alpha)
v_analitico= -A*w*np.sin(w*np.arange(0,(n+1))*dt+alpha)
a_analitico= -A*w**2*np.cos(w*np.arange(0,(n+1))*dt+alpha)

x=np.empty(n+1)
v=np.empty(n+1)
a=np.empty(n+1)

v[0]=v0
x[0]=x0
a[0]=a0

x_cromer=np.empty(n+1)
v_cromer=np.empty(n+1)
a_cromer=np.empty(n+1)

v_cromer[0]=v0
x_cromer[0]=x0
a_cromer[0]=a0


for i in range(0,n):    
    a[i]=-A*w**2*np.cos(w*i*dt+alpha)
    v[i+1]=v[i]+a[i]*dt
    x[i+1]=x[i]+v[i]*dt
    
    a_cromer[i]=-A*w**2*np.cos(w*i*dt+alpha)
    v_cromer[i+1]=v_cromer[i]+a_cromer[i]*dt
    x_cromer[i+1]=x_cromer[i]+v_cromer[i+1]*dt
    
a[i+1]=-A*w**2*np.cos(w*(i+1)*dt+alpha)
a_cromer[i+1]=-A*w**2*np.cos(w*(i+1)*dt+alpha)

plt.figure()
plt.plot(t,x,label='Euler')
plt.plot(t,x_analitico,label='Anal√≠tico')
plt.plot(t,x_cromer,label='Cromer')
# plt.plot(t,energia,label='W Res')
plt.legend()
plt.ylabel('x(m)')
plt.xlabel( 't(s)' )
plt.grid()

plt.figure()
plt.plot(t,v,label='Euler')
plt.plot(t,v_analitico,label='Anal√≠tico')
plt.plot(t,v_cromer,label='Cromer')
plt.legend()
# plt.plot(t,energia,label='W Res')
plt.ylabel('v(m/s)')
plt.xlabel( 't(s)' )
plt.grid()

plt.figure()
plt.plot(t,a,label='Euler')
plt.plot(t,a_analitico,label='Anal√≠tico')
plt.plot(t,a_cromer,label='Cromer')
plt.legend()
# plt.plot(t,energia,label='W Res')
plt.ylabel('a(m/s$^2$)')
plt.xlabel( 't(s)' )
plt.grid()

plt.show()
```
