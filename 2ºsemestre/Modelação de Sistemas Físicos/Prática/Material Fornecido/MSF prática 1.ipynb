**Como se determina o erro de quantidades que não se medem, mas que são funções de quantidades medidas?**
<br>

Exemplo: O perímetro e área de um rectângulo

Como se determina o erro de quantidades que não se medem, mas que são funções de quantidades medidas?
<br>
Adição e subtração de duas parcelas: largura, L, e comprimento, C.
<br>
$ S = L + C$
<br>
$ D = L - C$
<br>
Em que 
* $C = 3.0^+_- 0.1 \space cm$ 
* $L = 2.0^+_- 0.1 \space cm$ 
<br>

### Perímetro
<br>
Qual é o S e o dS? 
<br>
Qual é o D e o dD? 
<br>
Qual é o P e o pS?

```python
C = 3.0 # cm comprimento
dC = 0.1 # cm  erro comprimento
L = 2.0 # cm largura
dL = 0.1 #cm erro largura
S = C + L #quantidade soma
dS = dC +dL # erro absoluto da soma
D = C-L 
dD = dC +dL # erro absoluto da diferença
P = 2*C + 2*L # cálculo perimentro do rectângulo
dP = 2*dC+2*dL # cálculo erro absoluto  perimentro do rectângulo
print("Soma (S): {} cm".format(S))
print("dS: {} cm".format(dS))
print("Diferença (D): {} cm".format(D))
print("dD: {} cm".format(dD))
print("Perímetro: {} cm".format(P))
print("dP: {} cm".format(dP))
print("Perímetro Máximo: {} cm".format(P+dP))
print("Perímetro Mínimo: {} cm".format(P-dP))
```

### Área

Como se determina o erro de quantidades que não se medem, mas que são funções de quantidades medidas?
E o produto de 2 quantidades? Exemplo, a área do retângulo? 
<br>
<center><b>$ A = L * C$</b></center>

<br>
<center><b>$|\frac{\Delta A}{A}| = |\frac{\Delta L}{L}| +|\frac{\Delta C}{C}|  $ </b></center>


```python
A = C * L
dA = abs(dL/L + dC/C) * A #erro absoluto da área
print("Área Máxima: {} cm^2".format(A+dA))
print("Área Mínima: {} cm^2".format(A-dA))
print("dA: {} cm^2".format(dA))
```

# Algarismos Significativos

São os algarismos que se conhecem com certeza (100%) mais o 1$º$ algarismo que é afetado pelo erro

Ex:	Comprimento

a)	4.10 ± 0.02 m 		possui 3 algarismos significativos	(o erro afeta as centésimas)

b)	4.100 ± 0.02 m 		possui 3 algarismos significativos	(o erro afeta as centésimas)

c)	4.100 ± 0.2 m 		possui 2 algarismos significativos  (o erro afeta as décimas)

Permite escrever os valores de um modo mais simples: Escrever só os algarismos significativos

a)  4.10 m			 possui 3 algarismos significativos

b)  4.10 m			 possui 3 algarismos significativos

c)  4.1 m			 possui 2 algarismos significativos

<center><b>Em Física 4.10 m ≠ 4.1 m</b></center>

**Produto e divisão: O resultado da operação deve apresentar o número de algarismos significativos igual ao menor dos fatores**

Círculo de raio 

<center><b>$𝑟=6.0 ± 0,1 \space cm \space de \space área	$</b></center>

<center><b>$ 𝐴=𝜋 ×(6.0)^2 = 113.097⋯ cm^2$  </b></center>

apresenta-se com 2 algarismos significativos

<center><b>$ 𝐴=𝜋 ×(6.0)^2 = 1.1*10^2⋯ cm^2$  </b></center>

**Adição e subtração: O resultado da operação deve apresentar o número de casas decimais igual ao menor número  de casa decimais das parcelas.**
<center><b>$ 23.2 + 5.174 = 28.4$</b></center>
            
<center><b>$ 3.4 + 10 = 13$</b></center>
            
<center><b>$ 1,0001+0,0003=1,0004$ </b></center>
            
<center><b>$ 1,002−0,998=0,004 $</b></center>

Cálculos intermédios fazem-se com os todos os algarismos (na máquina de calcular ou computador)


```python
from math import log10, floor
import sigfig
import numpy as np

class SignificantAlg:
    def __init__(self, value, signif):
        """ Define as variáveis iniciais do objecto"""
        self._value = value # variavel protegida da class
        self._signif = signif
    
    def native_round(self, sg=None):
        """ Usa a função nativa do python round. 
        Return number rounded to ndigits precision after the decimal point. 
        If ndigits is omitted or is None, it returns the nearest integer to its input.
       
        https://docs.python.org/3/library/functions.html#round
        """
        if sg is not None:
            self._signif = sg
        return round(self._value, self._signif)
        
    def native_round_corrected(self):
        """A função anterior apresenta alguns erros no arrendodamento em alguns cenários.
        Por exemplo, sabemos que apenas o primeiro algarismo de uma medida de valor igual a 1234 é significativo. 
        Por forma a obter 1000, teria de se introduzir -3 como algarismos significativos por forma a obter solução correcta.
        """
        return round(self._value, self._signif-int(floor(log10(abs(self._value))))-1) 
    
    def sigfig_round(self):
        """(Usem este) Libraria python que apenas com a introdução da quantidade de números significativos corrige o valor da variável"""
        return sigfig.round(self._value, sigfigs = self._signif)
    
    def update_value_and_sig(self, value, signif):
        """Permite actualizar a variavel value e signif da class sem ter que iniciar uma nova class com novos valores"""
        self._value = value
        self._signif = signif
```

```python
v1 = 23.2 + 5.174 # 3 algarismos significativos
v2 = 3.4 + 10 # 2 algarismos significativos
v3 = 1.0001 + 0.0003 # 5 algarismos significativos
v4 = 1.002 - 0.998 # 4 algarismos significativos

signif_class = SignificantAlg(v1, 3)
v1_r1 = signif_class.sigfig_round()
print("V1: {}  V1_r: {}".format(v1, v1_r1))

signif_class.update_value_and_sig(v2, 2)
v2_r1 = signif_class.sigfig_round()
print("V2: {} V2_r: {}".format(v2, v2_r1))

signif_class.update_value_and_sig(v3, 5)
v3_r1 = signif_class.sigfig_round()
print("V3: {}  V3_r: {}".format(v3, v3_r1))

signif_class.update_value_and_sig(v4, 4)
v4_r1 = signif_class.sigfig_round()
print("V4: {}  V4_r:{}".format(v4,v4_r1))
```

Note: 1 milha = 1609 m

```python
def convert_miles_per_hour_to_km_per_hour(v):
    return v*1.609 # v-> valor da velocidade em milhas    1609 m ->1.609 km

velocity = 85.0 #milhas/hora
lim_speed = 50 # km/hora
velocity_km_per_hour = convert_miles_per_hour_to_km_per_hour(velocity) #converte milhas/h para km/h
if velocity_km_per_hour > lim_speed: #comparação das velocidades
    signif_class.update_value_and_sig(velocity_km_per_hour, 3) # actualiza a classe
    velocity_km_per_hour_r = signif_class.sigfig_round()# Algarismos significativos
    print("Viaja a {} km/h. Limite velocidade {} km/h".format(velocity_km_per_hour_r, lim_speed))

else:
    print("Tudo legal")
```

```python
def km_hour_to_m_s(s):
    return s/3.600 # 1 hora -> 3600 segundos  1 km -> 1000 m

velocity_m_s = km_hour_to_m_s(velocity_km_per_hour)
signif_class.update_value_and_sig(velocity_m_s, 3)
velocity_m_s = signif_class.sigfig_round()# Algarismos significativos
print("Velocity: {} m/s".format(velocity_m_s))
```

##  <font color='red'>Resolvido na aula prática 2 - Problema cap.1.5 -Regressão linear pelo método dos mínimos quadráticos</font> 

Quando se tem um conjunto 𝑥,𝑦 de N medições, o método dos mínimos  quadráticos oferece o ajuste linear que apresenta a menor diferença entre os valores medidos e os estimados por uma reta 𝑦=𝑚 𝑥+𝑏. Se se considerar que os erros que afetam os valores de 𝑦 são iguais, as expressões que o método fornece  são:


O coeficiente de determinação $𝑟^2$ é tal que quando ~1 indica um ótimo ajuste, enquanto que ~ 0 indica que o modelo não é linear. 



Escreva um programa em python que calcule as quantidades anteriores. Como teste ao programa escrito, compare os seus resultados com os de um problema conhecido, indicado na tabela abaixo, 



```python
import numpy as np
import matplotlib.pyplot as plt
import sigfig

class Regression:
    """Calcula a regressão linear de dois vectores de tamanhos iguais
    Poderão obtar por não programar em classes
    """
    def __init__(self, x_values=None, y_values=None):
        """Inicialização da classe e variáveis privadas"""
        self._x_values = x_values # valores x da função
        self._y_values = y_values # f(x)
        self._N = len(self._x_values) # número de valores do array
        self._sum_xy = None
        self._sum_x = None
        self._sum_y = None
        self._sum_pow_x_2 = None
        self._sum_pow_y_2 = None
        self._slope = None
        self._b_value = None
        self._err_slope = None
        self._err_b_value = None
        self._r_2 = None
        
    def sum_xy(self):
        """Somatório de x * y"""
        self._sum_xy = np.sum(self._x_values*self._y_values)
        return self._sum_xy
        
    def sum_x(self):
        """Somatório de x"""
        self._sum_x = np.sum(self._x_values)
        return self._sum_x
    
    def sum_y(self):
        """Somatório de y"""
        self._sum_y = np.sum(self._y_values)
        return self._sum_y
    
    def sum_pow_x_2(self):
        """Somatório do quadrado dos valores de x"""
        self._sum_pow_x_2 = np.sum(self._x_values**2)
        return self._sum_pow_x_2
    
    def sum_pow_y_2(self):
        """Somatório do quadrado dos valores de y"""
        self._sum_pow_y_2 = np.sum(self._y_values**2)
        return self._sum_pow_y_2
        
    def slope(self):
        """Cálculo do declive"""
        self._slope = (self._N*self.sum_xy()-self.sum_x()*self.sum_y())/(self._N*self.sum_pow_x_2()-self.sum_x()**2)
        return self._slope
        
    def b_value(self):
        """Cálculo da ordenada na origem"""
        self._b_value = (self.sum_pow_x_2()* self.sum_y() - self.sum_x()*self.sum_xy())/(self._N*self.sum_pow_x_2()-self.sum_x()**2)
        return self._b_value
    
    def  r_2(self):
        """Calculo do coefiente de determinação"""
        self._r_2 = (self._N*self.sum_xy()-self.sum_x()*self.sum_y())**2/((self._N*self.sum_pow_x_2()-self.sum_x()**2)*(self._N*self.sum_pow_y_2()-self.sum_y()**2))
        return self._r_2
    
    def err_slope(self):
        """Cálculo do erro do declive da recta"""
        self._err_slope = np.abs(self._slope)*np.sqrt(((1/self._r_2)-1)/(self._N-2))
        return self._err_slope
    
    def err_b_value(self):
        """Cálculo do erro da ordenada na origem"""
        self._err_b_value = self._err_slope*np.sqrt(self._sum_pow_x_2/self._N)
        return self._err_b_value
    
    def apply_regression(self):
        """Calcula todos os termos necessários para aplicar a regressão e imprime os resultados"""
        self.sum_xy()
        self.sum_x()
        self.sum_y()
        self.sum_pow_x_2()
        self.sum_pow_y_2()
        self.slope()
        self.b_value()
        self.r_2()
        self.err_slope()
        self.err_b_value()
        print("REGRESSION RESULTS")
        print("------------------")
        print("Sum_xy: {}".format(self.sum_xy()))
        print("Sum_x: {}".format(self.sum_x()))
        print("Sum_y: {}".format(self.sum_y()))
        print("sum_pow_x_2: {}".format(self.sum_pow_x_2()))
        print("sum_pow_y_2: {}".format(self.sum_pow_y_2()))
        print("slope (m): {} +- {}".format(self.slope(), self.err_slope()))
        print("b: {} +- {}".format(self.b_value(), self.err_b_value()))
        print("r^2: {}".format(self.r_2()))

## Funcionamento da classe
## x = ... (np.array)
## y = ...  (np.array)
## objecto_regressao = Regressao(x, y)
## objecto_regressao.apply_regression()
```
