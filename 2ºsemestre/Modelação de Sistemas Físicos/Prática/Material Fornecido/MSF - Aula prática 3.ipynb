*1. Um carro A segue numa estrada √† velocidade constante de 70 km/h onde o limite de velocidade √© de 40 km/h. Ao passar por um carro patrulha, este √∫ltimo parte em sua persegui√ß√£o √† acelera√ß√£o constante de 2,0 $m/s^2$.*

a)Fa√ßa o gr√°fico da lei do movimento do carro A e do carro patrulha, ùë•=ùë•(ùë°).

**Do enunciado podemos concluir que o carro A move-se num movimento rectil√≠neo uniforme pelo que**

$ v = \frac{\delta x}{\delta t} = constante$

$ \iff \delta x = v \delta y$

$ \iff \int_{x_0}^{x}\,dx\ = v \int_{0}^{t}\,dt\ $

$ \iff x - x_0 = v(t -0) $

$ \iff x = x_0 + vt $  

x -> Posi√ß√£o do carro A ao longo do tempo

**O carro patrulha move-se num movimento rectil√≠neo uniforme acelerado em acelera√ß√£o constante pelo que**

$ a = \frac{\delta v}{\delta t} = constante$

$ \iff \int_{v_0}^{v}\,dv\ = a \int_{t}^{0}\,dt\ $

$ \iff v - v_0 = a(t -0) $

$ \iff v = v_0 + at $  

Integrando a velocidade

$ v = \frac{\delta x}{\delta t}$

$ \iff \int_{x_0}^{x}\,dx\ = \int_{0}^{t}\,v_0 + at\, dt\ $

$ \iff x - x_0 = v_0t + \frac{1}{2}at^2 - 0$

$ \iff x = x_0 + v_0t + \frac{1}{2}at^2 $


```python
import numpy as np
import matplotlib.pyplot as plt

ap=2                        # m/s**2  acelera√ß√£o do carro patrulha
vx0a=70                     # km/h  velocidade inicial do carro A
vx0a=vx0a/3.6               # m/s   convers√£o ad velocidade paar m/s
vx0p=0
tempo=np.linspace(0,30,1000)    # cria√ß√£o de um vetor do tempo
xa=vx0a*tempo               # movimento uniforme acrro A
xp=0.5*ap*tempo**2          # movimento uniformente acelerado carro patrulha       

fig, ax = plt.subplots(1)
ax.set_xlabel( 'tempo (s)' )
ax.set_ylabel( 'x (m)' )
ax.plot(tempo,xa,label='Carro A')
ax.plot(tempo,xp,label='Carro Patrulha')
ax.set_title( 'Dois Carros' )
plt.legend()
plt.grid()
plt.savefig('DoisCarros.png',bbox_inches="tight")


```

```python
# symbolic calculus
import numpy as np # https://numpy.org/
import matplotlib.pyplot as plt # https://matplotlib.org/
import sympy as sym  # https://www.sympy.org/en/index.html
from numpy import block
```

```python
grav = 9.8 # acelera√ß√£o grav√≠tica
vt = 6.80  # velocidade do volante

# alinea a)
tem = np.linspace(0, 4, 110) # vector tempo com 110 valores entre os 0 e os 4 segundos
y = vt ** 2 / grav * np.log(np.cosh(grav * tem/ vt)) # equa√ß√£o do movimento do volante

fig, ax = plt.subplots(1)
ax.set_ylabel('y (m)')
ax.set_xlabel('t (s)')
plt.grid()
ax.plot(tem, y)
plt.show(block=False)

ax.set_title('Queda Volante com vy(0)=0')
plt.savefig('Quedavolante_y.png', bbox_inches="tight")
```

```python
t = sym.Symbol('t')
v = sym.Symbol('v')
vts = sym.Symbol('vt')
g = sym.Symbol('g')
D = sym.Symbol('D')
a = sym.Symbol('a')
aS = sym.Symbol('aS')

D = sym.Derivative(vts ** 2 / g * sym.log(sym.cosh(g / vts * t)), t, evaluate=True) # deriva a express√£o da posi√ß√£o
print('dy/dt=', D)
v = sym.simplify(D) # Simplifica a derivada anterior
print('(simplified) dy/dt =', v)
```
