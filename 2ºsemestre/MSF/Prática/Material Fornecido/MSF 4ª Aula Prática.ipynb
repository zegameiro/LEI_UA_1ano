```python
import numpy as np
import matplotlib.pyplot as plt

def position_and_velocity_eulerMethod(dt, tf, t0, g, v0y, y0):    
    n=np.int((tf-t0)/dt)
    print('n = ',n)

    t=np.zeros(n+1)
    vy=np.zeros(n+1)
    y=np.zeros(n+1)
  
    t[0]=t0
    vy[0]=v0y
    y[0]=y0

    for i in range(n): # Método de Euler
        t[i+1]=t[i]+dt
        vy[i+1]=vy[i]+g*dt       
        y[i+1]=y[i]+vy[i]*dt 
        
    return t, vy, y

```

b) Construa um programa que determine a velocidade do objeto, usando o método de Euler, no intervalo de tempo [0, 4 s]. Qual a velocidade em 3s?


```python
import numpy as np
# INPUT  Função Euler
def velocity_calculation_eulerMethod(dt, tf, t0, g, v0y):    
    n=np.int((tf-t0)/dt+0.1)
    print('Number of steps n = ',n)
    t=np.zeros(n+1)
    vy=np.zeros(n+1)
    t[0]=t0
    vy[0]=v0y   

    for i in range(n): # Método de Euler
        t[i+1]=t[i]+dt
        vy[i+1]=vy[i]+g*dt     
      
    return t, vy

dt=1    # Incremento do algoritmo. É mais preciso quando menor for o passo
tf=4.0    # Tempo final
t0=0      # Tempo inicial
g=9.80    # aceleração gravítica
v0y=0     # velocidade inicial


[tempo, velocidade] = velocity_calculation_eulerMethod(dt, tf, t0, g, v0y)
print(velocidade)

plt.plot(tempo, velocidade)
plt.xlabel ("tempo (s)")
plt.ylabel ("velocidade (m/s)")

print("A velocidade obtida pelo método de Euler: {} m/s".format(velocidade[abs(tempo-3)<0.0001][0]))
print("A velocidade teórica é igual a: {}".format(g*3)) # v=a*t

```

e) Construa um programa que determine a posição do objeto, usando o método de Euler, no intervalo de tempo [0, 3 s]. Qual a posição no instante 2 s, se o objeto partiu da posição 0 m? 


```python
import numpy as np
import matplotlib.pyplot as plt

def position_and_velocity_eulerMethod(dt, tf, t0, g, v0y, y0):    
    n=np.int((tf-t0)/dt)
    print('n = ',n)

    t=np.zeros(n+1)
    vy=np.zeros(n+1)
    y=np.zeros(n+1)
  
    t[0]=t0
    vy[0]=v0y
    y[0]=y0

    for i in range(n): # Método de Euler
        t[i+1]=t[i]+dt
        vy[i+1]=vy[i]+g*dt       
        y[i+1]=y[i]+vy[i]*dt 
        #y[i+1]=y[i]+vy[i+1]-g*dt
        
    return t, vy, y

p =3
dt=1   # Incremento do algoritmo. É mais preciso quando menor for o passo
tf=4.0    # Tempo final
t0=0      # Tempo inicial
g=9.80    # aceleração gravítica
v0y=0   # velocidade inicial
y0=0      # posição inicial

[tempo, velocidade, posicao] = position_and_velocity_eulerMethod(dt, tf, t0, g,v0y, y0)
# Para  verificar os valores
print(posicao[0:p])
print(velocidade[0:p])

tt = np.arange(0,4,dt)[0:p]
print(v0y*tt + 0.5*g*tt**2) #valor teórico da posição
print(v0y+g*tt) #valor teórico da velocidade
plt.plot(tempo, posicao)
plt.xlabel ("tempo (s)")
plt.ylabel ("posicao (m/s)")
p_2s_dt_1 = posicao[abs(tempo-2)<0.0001][0]

print("A posição do objecto obtida ao fim de 2 s pelo método de Euler: {} m".format(p_2s_dt_1))


```
