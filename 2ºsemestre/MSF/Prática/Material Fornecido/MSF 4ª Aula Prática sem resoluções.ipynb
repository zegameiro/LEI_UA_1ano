## Cap. 2 Movimento a 1 dimenso



Mtodo de Euler (mtodo numrico de integrao)
Programa Completo

```python
import numpy as np
import matplotlib.pyplot as plt

def position_and_velocity_eulerMethod(dt, tf, t0, g, v0y, y0):    
    n=np.int((tf-t0)/dt+0.1)
    print('n = ',n)

    t=np.zeros(n+1)
    vy=np.zeros(n+1)
    y=np.zeros(n+1)
  
    t[0]=t0
    vy[0]=v0y
    y[0]=y0

    for i in range(n): # Mtodo de Euler
        t[i+1]=t[i]+dt
        vy[i+1]=vy[i]+g*dt       
        y[i+1]=y[i]+vy[i]*dt 
        
    return t, vy, y

position_and_velocity_eulerMethod(1, 4, 0, 9.8, 0, 0)

```

b) Construa um programa que determine a velocidade do objeto, usando o mtodo de Euler, no intervalo de tempo [0, 4 s]. Qual a velocidade em 3s?


```python
import numpy as np

def velocity_methodEuler(dt, tf, t0, g, v0y, y0):
    n = int((tf - t0)/dt + 1)
    print("n = ", n)
    
    t = np.zeros(n+1)
    vy = np.zeros(n+1)
    
    t[0] = 0
    vy[0] = 0
    
    for i in range(n): 
        t[i+1]=t[i]+dt
        vy[i+1]=vy[i]+g*dt       
    
    return t, vy
    


t, vy = velocity_methodEuler(1, 4, 0, 9.8, 0, 0)
print("A velocidade aos 3 segundos  de ", round(vy[3],2), "m/s")
```

e) Construa um programa que determine a posio do objeto, usando o mtodo de Euler, no intervalo de tempo [0, 3 s]. Qual a posio no instante 2 s, se o objeto partiu da posio 0 m? 


```python
def position_methodEuler(dt, tf, t0, g, y0):
    n = int((tf - t0)/dt + 1)
    print("n = ", n)
    
    t = np.zeros(n+1)
    y = np.zeros(n+1)
    vy = np.zeros(n+1)
    
    t[0] = 0
    y[0] = 0
    
    for i in range(n): 
        t[i+1]=t[i]+dt
        vy[i+1]=vy[i]+g*dt
        y[i+1]=y[i]+vy[i+1]*dt       
    
    return t, vy, y

t, vy, y = position_methodEuler(1, 3, 0, 9.8, 0)
print("A posio no instante 2s  de", round(y[2],2),"m")

```
